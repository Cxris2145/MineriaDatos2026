## 1. Importación de Librerías y Configuración


```python
import warnings
import os

warnings.filterwarnings("ignore")
os.environ['PYTHONWARNINGS'] = 'ignore'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## 2. Comprensión de los Datos
En esta parte cargaremos los dos dataset ("ChileExport.csv" y "ecommerce_sales_data (2).csv")

```python
#CARGA INICIAL DEL LOS DATASETS

df_export = pd.read_csv("ChileExport.csv", sep=';', encoding='latin-1', decimal=',', on_bad_lines='skip', engine='python',quoting=3)
df_eco = pd.read_csv("ecommerce_sales_data (2).csv", encoding='utf-8-sig')
```

## 3. Preparación de los Datos
Detalle los pasos de limpieza y transformación de datos.

- Eliminación de nulos
- Conversión de categorías
- Normalización o estandarización

```python
# --- PREPARACIÓN EXPORTACIONES (Regresión) ---
# Usamos '1 category' con minúscula como viene en tu CSV
df_reg_clean = df_export[['Year', '1 category', 'Income in USD']].dropna()
df_reg_clean = pd.get_dummies(df_reg_clean, columns=['1 category'], drop_first=True)

X_reg = df_reg_clean.drop('Income in USD', axis=1)
y_reg = df_reg_clean['Income in USD']

# --- PREPARACIÓN E-COMMERCE (Market Basket) ---
df_eco.columns = df_eco.columns.str.strip()
df_eco['Product Name'] = df_eco['Product Name'].str.strip()
df_eco.dropna(axis=0, subset=['Order Date', 'Product Name'], inplace=True)

# Creamos la matriz de canasta usando 'Order Date' como identificador
basket_eco = (df_eco.groupby(['Order Date', 'Product Name'])['Quantity']
              .sum().unstack().reset_index().fillna(0)
              .set_index('Order Date'))

# Función para binarizar (1 si compró, 0 si no)
basket_encoded_eco = basket_eco.applymap(lambda x: 1 if x >= 1 else 0)

print(f"Preparación finalizada. Filas Exportación: {X_reg.shape[0]}, Canastas E-commerce: {basket_encoded_eco.shape[0]}")
```

#### 3.1 Depuración de Datos
- Manejo de valores nulos
- Detección y tratamiento de outliers
- Corrección de errores de tipeo

```python
# --- MODELO A: Regresión Lineal (Exportaciones) ---
X_train, X_test, y_train, y_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
modelo_reg = LinearRegression().fit(X_train, y_train)

# --- MODELO B: Market Basket Analysis (E-commerce) ---
# Soporte al 1% (0.01) para capturar patrones del nuevo dataset
frequent_itemsets_eco = apriori(basket_encoded_eco, min_support=0.01, use_colnames=True)
rules_eco = association_rules(frequent_itemsets_eco, metric="lift", min_threshold=1)
```

#### 3.2 Encoding (si aplica)
- One-Hot Encoding (Dataset de Exportaciones)
Para el modelo de regresión, se aplicó One-Hot Encoding a la variable 1 category utilizando la función pd.get_dummies.

- Encoding Binario (Dataset de E-commerce)
Para el análisis de Market Basket, se realizó una transformación hacia una Matriz Binarizada.

## 5. Modelado
En esta fase de modelado, el proyecto aborda dos problemas distintos. Para el dataset de exportaciones chilenas, se implementó un modelo de Regresión Lineal Múltiple (aprendizaje supervisado) con el fin de predecir el flujo de ingresos en dólares.

Para el dataset de e-commerce, se utilizó el Algoritmo Apriori para realizar un análisis de reglas de asociación (aprendizaje no supervisado).

## 6. Entrenamiento y Evaluación
Para el modelo de Regresión, se aplicó una técnica de validación mediante la división de datos (80% entrenamiento y 20% prueba), permitiendo medir el error del modelo con datos no vistos. Para el modelo de Asociación, se procesó la totalidad de la matriz transaccional para extraer reglas de comportamiento basadas en la frecuencia histórica de compra.


```python
# Resultados Regresión
y_pred = modelo_reg.predict(X_test)
print("="*40)
print(" RESULTADOS REGRESIÓN (CHILE EXPORT)")
print("="*40)
print(f"R2 Score: {r2_score(y_test, y_pred):.4f}")

# Resultados Market Basket
print("\n" + "="*55)
print(" TOP 10 REGLAS DE ASOCIACIÓN (E-COMMERCE)")
print("="*55)
if not rules_eco.empty:
    # Mostramos las mejores reglas por Lift
    display(rules_eco[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values('lift', ascending=False).head(10))
else:
    print("No se encontraron reglas. Baja el min_support a 0.005 en la sección de Modelado.")
```
